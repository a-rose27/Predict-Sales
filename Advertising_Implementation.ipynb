```python
# Function to handle outliers
def handle_outliers(data, method='remove'):
    for column in data.select_dtypes(include=[float, int]).columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        if method == 'remove':
            data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
        elif method == 'cap':
            data[column] = data[column].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))

    return data

# Remove outliers
data_no_outliers = handle_outliers(X, method='remove')

# Cap outliers
data_capped_outliers = handle_outliers(X, method='cap')
```

Cap the Outliers:
Replace the values that are below the lower bound with the lower bound and those above the upper bound with the upper bound.
